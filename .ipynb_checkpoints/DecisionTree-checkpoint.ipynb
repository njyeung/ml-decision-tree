```python
import numpy as np
import pandas as pd
```

```python
# information content, surprisal
def I(x):
    return -np.log2(x)

# entropy
def H(x):
    if x==0 or x==1:
        return 0
    return x * I(x)

def entropy(y):
    num_zero = 0
    num_one = 0

    for i in y:
        if i == 0:
            num_zero += 1
        if i == 1:
            num_one += 1
    
    probability_zero = num_zero / (num_zero + num_one)
    probability_one = num_one / (num_zero + num_one)

    return H(probability_zero) + H(probability_one)
```

```python
def find_split(df, X, y):
    best_threshold = None
    least_entropy = 99999
    column = None
    
    for x in X.columns:
        # find possible thresholds split
        unique_vals = df[x].sort_values().unique()
        thresholds = []
        for i in range(0, len(unique_vals)-1):
            thresholds.append((unique_vals[i] + unique_vals[i+1])/2)
    
        # try split and record entropy
        for threshold in thresholds:
            mask_left = X[x] <= threshold
            mask_right = X[x] > threshold
            
            df_left = X[mask_left]
            y_left = y[mask_left]
            
            df_right = X[mask_right]   
            y_right = y[mask_right]
            
            l = entropy(y_left)
            r = entropy(y_right)

            weighted_entropy = len(y_left)/len(y) * l + len(y_right)/len(y) * r
            
            if weighted_entropy < least_entropy:
                column = x
                best_threshold = threshold
                least_entropy = weighted_entropy
        
    return best_threshold, least_entropy, column
```

```python
def build_tree(df, X, y, depth=0):
    (best_threshold, least_entropy, column) = find_split(df, X, y)

    indent = "  " * depth
    
    if column is None or entropy(y) == 0:
        print(f"{indent}Leaf: class={y.iloc[0]}, samples={len(y)}")
        return

    print(f"{indent}Split on {column} <= {best_threshold:.2f} (entropy={least_entropy:.4f})")
        
    left_branch = df[df[column] <= best_threshold]
    right_branch = df[df[column] > best_threshold]
    
    if len(left_branch) > 0:
        build_tree(left_branch, left_branch[X.columns], left_branch[y.name], depth + 1)
    if len(right_branch) > 0:
        build_tree(right_branch, right_branch[X.columns], right_branch[y.name], depth + 1)

df = pd.read_csv('http://www.stat.wisc.edu/~jgillett/451/data/mtcars.csv', index_col=0)
# df = df.head(8)
df = df[['mpg', 'cyl', 'vs']]

X = df[['mpg', 'cyl']]
y = df['vs']

build_tree(df, X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import tree
import matplotlib.pyplot as plt

# Use the same data
df = pd.read_csv('http://www.stat.wisc.edu/~jgillett/451/data/mtcars.csv', index_col=0)
# df = df.head(8)
df = df[['mpg', 'cyl', 'vs']]

X = df[['mpg', 'cyl']]
y = df['vs']

# Fit sklearn decision tree
clf = DecisionTreeClassifier(criterion='entropy', random_state=0)
clf.fit(X, y)

plt.figure(figsize=(10,6))
plot_tree(clf, feature_names=X.columns, class_names=['V-engine', 'Straight'], filled=True)
plt.show()
```
